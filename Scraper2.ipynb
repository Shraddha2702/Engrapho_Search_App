```python
url = 'https://www.worldcat.org/search?qt=worldcat_org_all&q=Deep+Learning#%2528x0%253Abook%2Bx4%253Adigital%2529format'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
#name, author, year, source
```

```python
for each in soup.find_all(class_='name'):
    print(each.find('a').get_text())
    print(each.find('a')['href'])
```

```python
for one in soup.find_all(class_='author'):
    print(one.get_text()[3:])
```

```python
for one in soup.find_all(class_='publisher'):
    print(one.get_text()[11:])
    text = one.get_text()
    print(text[-5:].replace(' ', '').replace('.', '').replace(']',''))
```
