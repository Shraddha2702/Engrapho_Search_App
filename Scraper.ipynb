```python
import urllib.request
import time
```

```python
url = 'http://web.mta.info/developers/turnstile.html'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
one_a_tag = soup.find_all('a')[36]
link = one_a_tag['href']
```

```python
download_url = 'http://web.mta.info/developers/'+link
urllib.request.urlretrieve(download_url,'./'+link[link.find('/turnstile_')+1:]) 
```

```python
# To download the whole data set, let's do a for loop through all a tags
for i in range(36,len(soup.findAll('a'))+1): #'a' tags are for links
    one_a_tag = soup.findAll('a')[i]
    link = one_a_tag['href']
    download_url = 'http://web.mta.info/developers/'+ link
    urllib.request.urlretrieve(download_url,'./'+link[link.find('/turnstile_')+1:]) 
    time.sleep(1) #pause the code for a sec
```

```python
def download_images(output_folder, HEADERS, url):
    request = requests.get(url, HEADERS)
    soup = BeautifulSoup(request.content, 'html.parser')
    print(soup.status_code)
    #images = soup.findAll('img', {'class':'slide_image'})
    images = soup.findAll(class_='slide_image')
    print(images)
    i = 0
    for image in images:
        image_url = image.get('data-full').split('?')[0]
        print(image_url)
        urllib.request.urlretrieve(image_url, os.path.join(output_folder, 'ab'+str(i)+'.png'))
        i += 1
```
